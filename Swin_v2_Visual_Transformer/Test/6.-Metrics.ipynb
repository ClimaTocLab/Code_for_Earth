#### Open de PDF file

```python
path = os.path.join(os.getcwd())
pdf = PdfPages(os.path.join(path, '..', 'Metrics.pdf'))
```

```python
def plot_all_metrics(df):
    plt.figure(figsize=(10, 5))

    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan', 'lime', 'navy', 'gold', 'teal', 'indigo']
    markers = ['o', 'x', 's', 'd', '^', 'v', '<', '>', 'p', '*', 'h', 'H', '+', 'D', '|', '_', '1', '2', '3']


    columns = df.columns

    for i, col in enumerate(columns[1:]):
        plt.plot(np.array(df['Epoch']), np.array(df[col]), label=f'{col}', color=colors[i], marker=markers[i])

    # Añadir título y etiquetas
    plt.title('All metrics vs Epoch')
    plt.xlabel('Epoch')
    plt.ylabel('Metrics')
    plt.legend(fontsize=6)

    plt.grid(True)

    pdf.savefig()
```

```python
def plot_two_metrics(df, num1, num2):

    columns = df.columns
    
    var1 = columns[num1]
    var2 = columns[num2]

    plt.figure(figsize=(10, 5))

    plt.plot(np.array(df['Epoch']), np.array(df[var1]), label=f'{var1}', marker='o')

    plt.plot(np.array(df['Epoch']), np.array(df[var2]), label=f'{var2}', marker='o')

    plt.title(f'{columns[num1]} and {columns[num2]} vs Epoch')
    plt.xlabel('Epochs')
    plt.ylabel('Metrics')
    plt.legend()

    plt.grid(True)
    pdf.savefig()
```

```python
def comparison(train, valid, num):

    columns = train.columns
    
    var = columns[num]

    plt.figure(figsize=(10, 5))

    plt.plot(np.array(train['Epoch']), np.array(train[var]), label='train', marker='o')
    plt.plot(np.array(valid['Epoch']), np.array(valid[var]), label='valid', marker='o')

    #plt.ylim(0.3, 0.5)

    plt.title(f'{var} vs Epoch')
    plt.xlabel('Epochs')
    plt.ylabel(f'{columns[num]}')
    plt.legend()

    plt.grid(True)
    pdf.savefig()
```

```python
def comparison_all(train, valid):
    columns = train.columns

    # Crear la figura y el eje una sola vez
    fig, ax = plt.subplots(figsize=(10, 5))

    for col in columns[1:]:
        ax.plot(np.array(train['Epoch']), np.array(train[col]), label='train', marker='o')
        ax.plot(np.array(valid['Epoch']), np.array(valid[col]), label='valid', marker='o')
        ax.set_title(f'{col} vs Epoch')
        ax.set_xlabel('Epochs')
        ax.set_ylabel(f'{col}')
        ax.grid(True)
        ax.legend()

        display(fig)  # Muestra la gráfica en el notebook
        time.sleep(1)  # Espera 1 segundos

        if col != columns[-1]:
            clear_output(wait=True)  # Limpia la salida de la celda
            ax.cla()  # Limpia el contenido del eje       
```

```python
plot_all_metrics(df_1)
```

```python
columns = df_1.columns
for i, col in enumerate(columns):
    print(i, col)
```

```python
for i in range(1, df_1.shape[1], 2):
    if i + 1 < df_1.shape[1]:
        plot_two_metrics(df_1, i, i + 1)
    else:
        plot_two_metrics(df_1, i, i)
```

```python
plot_all_metrics(df_2)
```

```python
columns = df_2.columns
for i, col in enumerate(columns):
    print(i, col)
```

```python
for i in range(1, df_2.shape[1], 2):
    if i + 1 < df_2.shape[1]:
        plot_two_metrics(df_2, i, i + 1)
    else:
        plot_two_metrics(df_2, i, i)
```

```python
for i in range(1, df_1.shape[1]):
    comparison(df_1, df_2, i)
```

```python
comparison_all(df_1, df_2)
```
